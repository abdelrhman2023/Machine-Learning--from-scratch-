# Booleans (Logical Values)
In programming, you often need to know if an expression is **true** or **false**.<p>
You can evaluate any expression in R, and get one of two answers, `TRUE` or `FALSE`.<p>
When you compare two values, the expression is evaluated and R returns the logical answer:

```r
10 > 9    # TRUE because 10 is greater than 9
10 == 9   # FALSE because 10 is not equal to 9
10 < 9    # FALSE because 10 is greater than 9
```

# Operators
Operators are used to perform operations on variables and values.

## R Arithmetic Operators
**Arithmetic operators are used with numeric values to perform common mathematical operations:**

`Operator	   Name	                          Example`<p>
`+            Addition	                         x + y`<p>
`-	        Subtraction	                      x - y`<p>
`*	        Multiplication                       x * y`<p>
`/	        Division	                         x / y`<p>
`^	        Exponent	                         x ^ y`<p>
`%%	       Modulus(Remainder from division)     x %% y `<p>
`%/%	      Integer Division                     x%/%y   `<p>

## R Comparison Operators
**Comparison operators are used to compare two values:**

`Operator	    Name    	              Example`

` ==	         Equal	                   x == y`

` !=	         Not equal	               x != y`

` >	          Greater than	            x > y`

` <	          Less than	               x < y`

` >=	         Greater than or equal to    x >= y`

` <=	         Less than or equal to       x <= y`

## R Logical Operators
**Logical operators are used to combine conditional statements:**

`Operator	   Description`<p>
`&	          Element-wise Logical AND operator. It returns TRUE if both elements are TRUE`<p>
`&&	         Logical AND operator - Returns TRUE if both statements are TRUE`<p>
`|	          Elementwise- Logical OR operator. It returns TRUE if one of the statement is TRUE`<p>
`||	         Logical OR operator. It returns TRUE if one of the statement is TRUE.`<p>
`!	          Logical NOT - returns FALSE if statement is TRUE`<p>

## R Miscellaneous Operators
**Miscellaneous operators are used to manipulate data:**

`Operator	       Description	                               Example`<p>
`:	              Creates a series of numbers in a sequence	 x <- 1:10`<p>
`%in%	           Find out if an element belongs to a vector	x %in% y`<p>
`%*%	            Matrix Multiplication	                     x <- Matrix1 %*% Matrix2`<p>
